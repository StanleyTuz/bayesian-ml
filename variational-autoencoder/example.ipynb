```python
import numpy as np 
import matplotlib.pyplot as plt 
import scipy.stats as st

```

Generate some example data. Mixture of two Gaussians seems like a straightforward example.


```python
mu_exact = [0, 7]
sig_exact = [3, 1]
p_mix = 0.5 # Gaussian mixture parameter

n_pts = 400 


def data_gen(mu, sigmas, phi, n_pts):
    y = []
    for i in range(n_pts):
        idx = np.random.binomial(1, phi, 1) # draw a 0 or 1 from binomial (coin flip)
        if idx:
            y.append(st.norm(mu[0], sigmas[0]).rvs()) # sample from the first Gaussian component
        else:
            y.append(st.norm(mu[1], sigmas[1]).rvs()) # sample from the second component
    return(np.array(y))

y_data = data_gen(mu_exact, sig_exact, p_mix, n_pts)

# get the actual Gaussians - we know these from the parameters

x = np.linspace(-10,12)

plt.hist(y_data, 30, density=True, alpha=0.5)
plt.plot(x, st.norm(mu_exact[0], sig_exact[0]).pdf(x), color='r')
plt.plot(x, st.norm(mu_exact[1], sig_exact[1]).pdf(x), color='b')
plt.show()
```
