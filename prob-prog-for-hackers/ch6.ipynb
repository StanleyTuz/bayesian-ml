```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(12.5, 3))

colors = ["#348ABD", "#A60628", "#7A68A6", "#467821"]

x = np.linspace(0,1)
y1, y2 = stats.beta.pdf(x, 1,1), stats.beta.pdf(x, 10,10)

p = ax.plot(x, y1, 
    label='An objective prior \n(uninformative, \n"Principle of Indifference")')
ax.fill_between(x, 0, y1, color = p[0].get_color(), alpha = 0.3)

p = ax.plot(x,y2 ,
     label = "A subjective prior \n(informative)")
ax.fill_between(x, 0, y2, color = p[0].get_color(), alpha = 0.3)

p = ax.plot(x[25:], 2*np.ones(25), label = "another subjective prior")
ax.fill_between(x[25:], 0, 2, color = p[0].get_color(), alpha = 0.3)

ax.set_ylim(0,4)

# plt.ylim(0, 4)
leg = ax.legend(loc = "upper left")
leg.get_frame().set_alpha(0.4)
ax.set_title("Comparing objective vs. subjective priors for an unknown probability");
```

>In instances of scientific research, the choice of an objective prior [over a subjective prior] is obvious.

>The significant difference between [frequentists and Bayesian methods] is that Bayesian methods have a prior distribution, with hyperparameters Î±, while empirical methods do not have any notion of a prior.

Empirical Bayes: use frequentist methods to select $\alpha$, and use Bayesian methods on the original problem.

### Gamma Distribution

Distribution over $\mathbb{R}^+$

```python
fig, ax = plt.subplots(figsize=(12.5, 5))

params = [(1, 0.5), (9, 2), (3, 0.5), (7, 0.5)]
x = np.linspace(0.001, 20, 150)
for alpha, beta in params:
    y = stats.gamma.pdf(x, alpha, scale=1./beta)
    lines = plt.plot(x, y, label=f"({alpha:.1f}, {beta:.1f})", lw=3)
    ax.fill_between(x, 0, y, alpha=0.2, color=lines[0].get_color())
    ax.autoscale(tight=True)

ax.legend(title=r"$\alpha, \beta$ - parameters")
plt.show()
```

### Beta Distribution

Distribution over $\left[0,1\right]$.

```python
fig, ax = plt.subplots(figsize=(12.5, 5))

params = [(2, 5), (1, 1), (0.5, 0.5), (5, 5), (20, 4), (5, 1)]
x = np.linspace(0.01, 0.99, 100)
for alpha, beta in params:
    y = stats.beta.pdf(x, alpha, beta)
    lines = plt.plot(x, y, label=f"({alpha}, {beta})", lw=3)
    ax.fill_between(x, 0, y, alpha=0.2, color=lines[0].get_color())
    ax.autoscale(tight=True)

ax.legend(title=r"$\left(\alpha, \beta\right)$ - parameters", loc="upper left")
plt.show()
```

Beta is the conjugate prior for the Binomial distribution (the product is Beta-distributed).
